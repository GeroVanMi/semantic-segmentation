```python
display(Image.open('../data/images/beagle_10.png'))
```

```python
file_name = '../data/masks/beagle_10.png'

im = Image.open(file_name) # Can be many different formats.
pix = im.load()

print(im.size)
array = []

for y in range(0, im.size[1]):
    current_row = []
    for x in range(0, im.size[0]):
        value = pix[x, y]
        current_row.append(value)
        if value == 1:
            pix[x, y] = 128
        if value == 2:
            pix[x, y] = 255
    array.append(current_row)

np_array = np.asarray(array)
np_array
```
